```python
import pandas as pd

# Créez vos DataFrames
df_codes_postaux = pd.read_csv("Export_Données_Fournisseurs_MontantFactures.csv",sep=';')

df_departements_regions = pd.read_csv("regions_et_departements_francais.csv",sep=';')

# Extrait les deux premiers chiffres du code postal
df_codes_postaux['CodePostalPrefix'] = df_codes_postaux['shipping_postalcode'].str[:2]

# Fusionne les DataFrames sur les deux premiers chiffres du code postal
df_resultat = pd.merge(df_codes_postaux, df_departements_regions, left_on='CodePostalPrefix', right_on='NuméroDépartement', how='left')

# Affiche le résultat
print(df_resultat[['shipping_postalcode', 'Région']])
```
